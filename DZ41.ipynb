```python
import pandas as pd
```

```python
DATA_PATH = 'names/'
```

```python
def count_top3(years):
    names_all = pd.DataFrame()
    for year in years:
        names = pd.read_csv(DATA_PATH + 'yob'+str(year)+'.txt', names=['Name','Gender','Count'])
        names_all = pd.concat([names_all, names], names=['Year', 'Pos'])
    return list(names_all.groupby('Name').sum().sort_values(by='Count', ascending=False).head(3).reset_index()['Name'])
```

```python
count_top3([1880])
```

```python
count_top3([1900, 1950, 2000])
```
